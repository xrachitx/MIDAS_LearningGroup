```python
from sklearn.preprocessing import *
import matplotlib.pyplot as plt
import pandas as pd
import numpy
# from sklearn import linear_model
from sklearn.linear_model import LogisticRegression
```

```python
def prepareData(train,test):
    takeCols = ["Survived","Pclass","Sex","Age","SibSp","Parch","Fare","Embarked"]
    trainDf = pd.read_csv(train, usecols= takeCols)
    testDf = pd.read_csv(test,usecols= takeCols[1:])
    trAgeMean = trainDf["Age"].mean()
    print(trAgeMean)
    trainDf["Age"]= trainDf["Age"].fillna(trainDf["Age"].mean())
    testDf["Age"]= testDf["Age"].fillna(testDf["Age"].mean())
    trainDf["Embarked"]=trainDf["Embarked"].fillna("navail")
    testDf["Embarked"]= testDf["Embarked"].fillna("navail")
    testDf["Fare"] = testDf["Fare"].fillna(testDf["Fare"].mean())
    for i in range(trainDf.shape[0]):
        if trainDf["Sex"].iloc[i] =="male":
            trainDf["Sex"].iloc[i] = 0
        else:
            trainDf["Sex"].iloc[i] = 1
    d = {}
    cnt = 0
    for i in range(trainDf.shape[0]):
        if trainDf["Embarked"].iloc[i] in d.keys():
            trainDf["Embarked"].iloc[i] = d[trainDf["Embarked"].iloc[i]]
        else:
            cnt +=1
            d[trainDf["Embarked"].iloc[i]] = cnt
            trainDf["Embarked"].iloc[i] = d[trainDf["Embarked"].iloc[i]]
    for i in range(testDf.shape[0]):
        if testDf["Sex"].iloc[i] =="male":
            testDf["Sex"].iloc[i] = 0
        else:
            testDf["Sex"].iloc[i] = 1
    for i in range(testDf.shape[0]):
        if testDf["Embarked"].iloc[i] in d.keys():
            testDf["Embarked"].iloc[i] = d[testDf["Embarked"].iloc[i]]
        else:
            cnt +=1
            d[testDf["Embarked"].iloc[i]] = cnt
            testDf["Embarked"].iloc[i] = d[testDf["Embarked"].iloc[i]]
    print(d)
    train = trainDf.values
    test = testDf.values
    min_max_scaler = MinMaxScaler()
    X_Train = train[:,1:]
    X_Train = min_max_scaler.fit_transform(X_Train)
    X_Train = pd.DataFrame(X_Train)
    Y_Train = train[:,0]
    Y_Train.reshape(891,-1)
    Y_Train = pd.DataFrame(Y_Train)
#     print(Y_Train)
    for i in range((Y_Train.shape[0])):
        Y_Train[0].iloc[i]= int(Y_Train[0].iloc[i])
    X_Test = test[:,:]
    X_Test = min_max_scaler.fit_transform(X_Test)
    X_Test = pd.DataFrame(X_Test)
    return X_Train, Y_Train, X_Test
```
